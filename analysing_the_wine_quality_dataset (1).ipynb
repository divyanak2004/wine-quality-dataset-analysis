## Importing librarieas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import seaborn as sns
```

## Load the dataset

```python
white = pd.read_csv("winequality-white.csv",sep = ";")
red = pd.read_csv("winequality-red.csv",sep = ";")
```

## Clean and prepare the data.
 

```python
# No. of samples and columns in each dataset

print(white.shape)
print(red.shape)
```

```python
# unique values of quality in red wine dataset
print(red.quality.nunique())

# unique values of quality in white wine dataset
print(white.quality.nunique())
```

```python
# Check for missing values in the Red Wine dataset
print(red.isnull().sum())

# Check for missing values in the White Wine dataset
print(white.isnull().sum())
```

```python
# Check for any duplicate rows in red wine dataset
print(red.duplicated().sum())
```

```python
# Check for any duplicate rows in white wine dataset
print(white.duplicated().sum())
```

```python
# Remove the duplicate rows
red.drop_duplicates(inplace=True)
white.drop_duplicates(inplace=True)
```

```python
# Check for any duplicate rows in red wine dataset
print(red.duplicated().sum())
```

```python
# Check for any duplicate rows in white wine dataset
print(white.duplicated().sum())
```

```python
# create color array for red dataframe
color_red = np.repeat('red',red.shape[0])
color_white = np.repeat('white',white.shape[0])
red['color'] = color_red
white['color'] = color_white

# append dataframes
df = pd.concat([red, white], ignore_index=True)
df
```

## Explore the data using visualizations and statistical analyses

```python
# distribution of the wine quality scores
sns.histplot(data=df, x='quality')
plt.show()
```

```python
# heatmap to visualize the correlation between the different features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Create subplots for histograms of selected columns
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='total sulfur dioxide', hue="color", kde=True, bins=30)
plt.title('Histogram of Total Sulfur Dioxide')
plt.xlabel('total sulfur dioxide')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='pH', hue="color", kde=True, bins=30)
plt.title('Histogram of pH')
plt.xlabel('pH')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='alcohol', hue="color", kde=True, bins=30)
plt.title('Histogram of Alcohol')
plt.xlabel('Alcohol')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='fixed acidity', hue="color", kde=True, bins=30)
plt.title('Histogram of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
```

```python
#Which type of wine (red or white) associated with higher quality
df.groupby(by="color")["quality"].mean()
```

```python
#Do wines with higher alcoholic content receive better ratings
df.alcohol.median()
```

```python
# select samples with alcohol content less than the median
low_alcohol = df.query('alcohol < 10.3')
high_alcohol = df.query('alcohol >= 10.3')
num_samples = df.shape[0]
num_samples == low_alcohol['quality'].count() + high_alcohol['quality'].count()
```

```python
print(low_alcohol['quality'].mean())
print(high_alcohol['quality'].mean())
```

```python
mean_quality_low = low_alcohol['quality'].mean()
mean_quality_high = high_alcohol['quality'].mean()
```

```python
locations = [1, 2]
heights = [mean_quality_low, mean_quality_high]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating')
```

Wines with higher alcohol content received higher rating
